```python
def rescaling(sx, sy, img):
    # Scale height (axis=0)
    img = np.repeat(img, sy, axis=0)
    # Scale width (axis=1)
    img = np.repeat(img, sx, axis=1)
    return img


arr = np.array([[[199, 208, 211],
                 [250, 252, 252]],
                [[255, 255, 255],
                 [255, 255, 255]]])

rescaling(2, 2, arr)
```
