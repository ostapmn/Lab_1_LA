```python
from turtle import width


def get_scaling_matrix(scale_factor_x, scale_factor_y):
  return np.array([[scale_factor_x, 0],
                   [0, scale_factor_y]], dtype=np.float32)

def rescaling(sx, sy, img):
    # Scale height (axis=0)
    img = np.repeat(img, sy, axis=0)
    # Scale width (axis=1)
    img = np.repeat(img, sx, axis=1)
    return img


def rescaling_with_matrix(sx, sy, img):
    M = get_scaling_matrix(sx, sy)
    height, width = img.shape[:2]
    new_width = int(width * M[0, 0])
    new_height = int(height * M[1, 1])
    scaled_img = np.zeros((new_height, new_width, img.shape[2]), dtype=img.dtype)
    for i in range(height):
       for j in range(width):
            scaled_img[i*sy:(i+1)*sy, j*sx:(j+1)*sx] = img[i, j]
    
    return scaled_img

def get_vertical_shear_matrix(factor):
  return np.array([[1, 0], [factor, 1]])

def apply_sheering(factor, img):
    M = get_vertical_shear_matrix(factor)
    height, width = img.shape[:2]
    new_height = height + int(abs(factor) * width)
    sheared_img = np.zeros((new_height, width, img.shape[2]), dtype=img.dtype)
    for i in range(height):
        for j in range(width):
                new_i = i + int(factor * j)
                if 0 <= new_i < new_height:
                    sheared_img[new_i, j ] = img[i, j]
    return sheared_img

arr = np.array([[[199, 208, 211],

                 [250, 252, 252]],
                [[255, 255, 255],
                 [255, 255, 255]]])
rescaling_with_matrix(2, 2, arr)
```
